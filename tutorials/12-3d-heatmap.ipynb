# Three-Dimensional Heatmaps

ComplexHeatmap provides `Heatmap3D` for 3D bar-style heatmaps and `bar3D`
for drawing individual 3D bars. The 3D effect is a projection where each
cell has a bar whose height (length in the z-direction) corresponds to its
value. The `theta` parameter controls the viewing angle.

`Heatmap3D` accepts almost all arguments from `Heatmap`, so clustering,
splitting, and annotations all work. However, 3D visualization is best kept
simple and applied to small matrices.

```python
import numpy as np
import pandas as pd
import complexheatmap as ch
from complexheatmap import *
import grid_py as gp
```

## Preparing a small demo matrix

`Heatmap3D` requires non-negative values (bar height must be >= 0).
We use `runif`-style values in [0, 1].

```python
np.random.seed(42)
mat = np.random.rand(5, 6)
print(f"Matrix shape: {mat.shape}")
print(f"Range: [{mat.min():.3f}, {mat.max():.3f}]")
```

## Basic Heatmap3D

The simplest call creates a 3D heatmap with default settings. Clustering
is applied but dendrograms are hidden by default.

```python
ht3d = Heatmap3D(mat)
print(f"Type: {type(ht3d).__name__}")
print(f"Is Heatmap3D: {isinstance(ht3d, Heatmap3D)}")
```

```python
ht3d
```

## Heatmap3D with row and column names

```python
ht3d_named = Heatmap3D(
    mat,
    row_names=[f"row{i}" for i in range(5)],
    column_names=[f"col{j}" for j in range(6)],
    title="Named 3D",
)
ht3d_named
```

## Heatmap3D with custom color function

A `color_ramp2` function can be passed via the `col` argument, just like
a regular `Heatmap`.

```python
col_fun = color_ramp2([0, 0.5, 1], ["blue", "white", "red"])

ht3d_col = Heatmap3D(mat, col=col_fun)
ht3d_col
```

## Varying the theta angle

The `theta` parameter controls the projection angle of the 3D bars.
It ranges from 0 to 90 degrees. A larger theta pushes the bars more
to the right and up.

```python
for angle in [20, 45, 60, 80]:
    ht = Heatmap3D(mat, theta=float(angle))
    print(f"theta={angle}: construction OK = {isinstance(ht, Heatmap3D)}")
```

## Heatmap3D with a wider range of values

Let's try a 10x10 matrix with named rows and columns, and a custom
color mapping.

```python
np.random.seed(7)
mat10 = np.random.rand(10, 10)
row_names = [chr(65 + i) for i in range(10)]  # A-J
col_names = [chr(97 + i) for i in range(10)]  # a-j

ht3d_big = Heatmap3D(
    mat10,
    row_names=row_names,
    column_names=col_names,
    title="This is a 3D heatmap",
    col=color_ramp2([0, 0.5, 1], ["white", "orange", "darkred"]),
)
ht3d_big
```

## The `bar3D` function

`bar3D` is the low-level function that draws individual 3D bars as grid
grobs. It can be used directly for custom annotations or standalone plots.

```python
grob = bar3D(mat)
grob
```

## bar3D with a color function

```python
col_fun = color_ramp2([0, 0.5, 1], ["blue", "white", "red"])
grob_col = bar3D(mat, col=col_fun)
print(f"bar3D with col_fun: type = {type(grob_col).__name__}")
```

## bar3D with names and title

```python
grob_named = bar3D(
    mat,
    row_names=[f"r{i}" for i in range(5)],
    column_names=[f"c{j}" for j in range(6)],
    title="Bar 3D",
)
print(f"bar3D named: type = {type(grob_named).__name__}")
```

## bar3D with non-negative matrix

Ensure that taking absolute values of a matrix with negatives also works.

```python
mat_neg = np.random.randn(5, 6)
mat_pos = np.abs(mat_neg)
grob_pos = bar3D(mat_pos)
print(f"bar3D with abs values: type = {type(grob_pos).__name__}")
```

## Heatmap3D with color_ramp2 using diverging colors

```python
col_fun_div = color_ramp2([-2, 0, 2], ["blue", "white", "red"])
ht3d_div = Heatmap3D(mat, col=col_fun_div)
print(f"Diverging color Heatmap3D: {isinstance(ht3d_div, Heatmap3D)}")
```

## Summary

This tutorial covered the 3D visualization capabilities:

- `Heatmap3D(mat)`: creates a 3D bar-style heatmap
  - Accepts `row_names`, `column_names`, `title`, `col`, `theta`
  - Requires non-negative input values
  - Supports all standard `Heatmap` arguments (clustering, splitting, etc.)
- `bar3D(mat)`: low-level function for drawing 3D bar grobs
  - Accepts `col`, `row_names`, `column_names`, `title`
  - Returns a grid grob for composition
- The `theta` parameter (0-90) controls the 3D projection angle
- Best used on small matrices for visual clarity
